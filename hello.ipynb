<H1>Data Manipulation using RDD</H1>

```python
# Import modules
from pyspark import SparkContext
```

```python
# Create instance for Sparkcontext
sc = SparkContext()
```

```python
# Load dataset as RDD
filePath = "Nutritional Facts for Most Common Foods/nutrients_csvfile.csv"
dataSetRDD = sc.textFile(filePath)
print("Data type is ", type(dataSetRDD))
```

```python
# Basic RDD Transformation
# Protein greater than 30

# splits csv into list
dataSetRDD_map = dataSetRDD.map(lambda line: line.split(","))

# exclude the title
header = dataSetRDD_map.first()
filteredData = dataSetRDD_map.filter(lambda x: x!=header)

# extract a column
food_name_index = 0
protein_val_index = 4 
dataSetRDD_filtered = filteredData.map(lambda row: (row[food_name_index], row[protein_val_index]))
dataSetRDD_filtered_collected = dataSetRDD_filtered.collect()

# print output
print(header[food_name_index], header[protein_val_index])
for val in dataSetRDD_filtered_collected:
    print(val)
```

```python
# RDD Actions
dataSetRDD_title = dataSetRDD.first()
dataSetRDD_title_list = dataSetRDD_title.split(',')
# dataSetRDD_title = dataSetRDD_title.collect()
dataSetRDD_mapped = dataSetRDD.map(lambda title: title.split(','))
print(dataSetRDD_title_list)
print(type(dataSetRDD_mapped))
# print(dataSetRDD.collect())
# print(dataSetRDD.take(2))
```
